```python
#2 — Imports
from pathlib import Path
import os
from dotenv import load_dotenv
from tqdm.auto import tqdm
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
#3 – Load Environment
load_dotenv()

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

assert GEMINI_API_KEY is not None
```

```python
#4 – Paths
PROJECT_ROOT = Path("..").resolve()

ARTIFACTS = PROJECT_ROOT / "artifacts"

INPUT_FILE = ARTIFACTS / "chunk_documents.pkl"

OUTPUT_FILE = ARTIFACTS / "vector_records.pkl"
```

```python
import pickle
#5 – Load Chunk Documents
with open(INPUT_FILE, "rb") as f:
    chunk_documents = pickle.load(f)

print(f"Chunks : {len(chunk_documents)}")
```

```python
#6 – Initialize Embedding Model
embedding_model = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001",
    output_dimensionality=768
)
```

```python
#7 – Test Embedding
vector = embedding_model.embed_query(
    "How do I reset my password?"
)

print(len(vector))
```

```python
#8 – Prepare Texts
texts = [
    doc.page_content
    for doc in chunk_documents
]

len(texts)
```

```python
#9 – Batch Embedding
BATCH_SIZE = 100

embeddings = []

for start in tqdm(range(0, len(texts), BATCH_SIZE)):

    batch = texts[start:start+BATCH_SIZE]

    vectors = embedding_model.embed_documents(batch) 

    embeddings.extend(vectors)
```

```python
#10 – Verify
print(len(embeddings))
print(len(embeddings[0]))
```

```python
#11 – Create Vector Records
vector_records = []

for doc, embedding in zip(chunk_documents, embeddings):

    record = {

        "id": doc.metadata["chunk_id"],

        "values": embedding,

        "metadata": doc.metadata,

        "text": doc.page_content,

        "embedding_model": "gemini-embedding-001",

        "embedding_dimension": len(embedding)

    }

    vector_records.append(record)
```

```python
#12 – Inspect Record
vector_records[0]
```

```python
#13 – Save Records
with open(OUTPUT_FILE, "wb") as f:
    pickle.dump(vector_records, f)

print(f"Saved {len(vector_records)} records")
```

```python
#14 – Reload Test
with open(OUTPUT_FILE, "rb") as f:
    loaded = pickle.load(f)

print(type(loaded))

print(type(loaded[0]))
```

```python
#15 – Summary
print("="*60)

print("EMBEDDING GENERATION COMPLETE")

print("="*60)

print(f"Records : {len(vector_records)}")

print(f"Embedding Model : gemini-embedding-001")

print(f"Dimensions : {len(vector_records[0]['values'])}")

print(f"Output : {OUTPUT_FILE}")
```
