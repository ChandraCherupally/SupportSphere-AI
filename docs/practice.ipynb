```python
from google import genai
from langchain_google_genai import ChatGoogleGenerativeAI
import os
CLOUD_API_KEY = os.getenv("GOOGL_CLOUD_API_KEY")
API_KEY = os.getenv("GOOGL_API_KEY")
```

```python
client = genai.Client(vertexai=False,api_key=API_KEY)
response = client.models.generate_content(model="gemini-3.5-flash", contents="Explain how quantum computing works in one short paragraph.")
print(response.text)
```

```python
client = genai.Client(vertexai=True,api_key=CLOUD_API_KEY)
response = client.models.generate_content(model="gemini-3.5-flash",contents="Explain how quantum computing works in one short paragraph.")
print(response.text)
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3.5-flash",api_key=API_KEY,vertexai=False,)
response = llm.invoke([{"role": "user", "content": "Explain machine learning in one sentence."}])
print(response.content)
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-3.5-flash", 
    api_key=CLOUD_API_KEY, 
    vertexai=True,
)

response = llm.invoke([{"role": "user", "content": "Explain machine learning in one sentence."}])
print(response.content)
```
